```python
import tensorflow as tf
# import keras
import os
from keras_radam import RAdam
# from keras import backend as K

import tensorflow as tf
from tensorflow.python import keras
import tensorflow.keras.backend as K

from keras_bert import Tokenizer
from keras_bert import get_custom_objects
from keras_bert import load_trained_model_from_checkpoint

import numpy as np
```

```python
SEQ_LEN = 128
BATCH_SIZE = 16
EPOCHS = 3
LR = 1e-5
```

```python
pretrained_path = 'multi_cased_L-12_H-768_A-12'
config_path = os.path.join(pretrained_path, 'bert_config.json')
checkpoint_path = os.path.join(pretrained_path, 'bert_model.ckpt')
vocab_path = os.path.join(pretrained_path, 'vocab.txt')
```

```python
model = load_trained_model_from_checkpoint(
    config_path,
    checkpoint_path,
    trainable=True,
    training = False,
    seq_len=SEQ_LEN,
    output_layer_num=4
)
```

```python
import keras_bert
from keras.layers import Dense, Input, Flatten,SpatialDropout1D
from keras.layers import Conv1D, MaxPooling1D, Embedding, Concatenate, Dropout,GlobalMaxPool1D,Lambda
from keras.models import Model
from keras.layers import Bidirectional,LSTM,GRU
from keras.optimizers import Adam
```

```python
import codecs
token_dict = {}
with codecs.open(vocab_path, 'rb','utf-8') as reader:
    for line in reader:
        token = line.strip()
        token_dict[token] = len(token_dict)
```

```python
from keras_bert import Tokenizer
tokenizer = Tokenizer(token_dict,cased=True)
```

```python
from keras_bert.layers import Extract
from keras.regularizers import l1
```

```python
from keras import optimizers
```

```python
inputs = model.inputs[:2]
newout = Extract(index=0)(model.output)
newout = keras.layers.Dense(128,activation='relu')(newout)
outputs = keras.layers.Dense(units=1, activation='sigmoid')(newout)
model = keras.models.Model(inputs, outputs)
model.compile(
  optimizer= RAdam(lr=LR),
  loss='binary_crossentropy',
  metrics=['accuracy'],
)
```

```python
sess = K.get_session()
uninitialized_variables = set([i.decode('ascii') for i in sess.run(tf.report_uninitialized_variables())])
init_op = tf.variables_initializer(
    [v for v in tf.global_variables() if v.name.split(':')[0] in uninitialized_variables]
)
sess.run(init_op)
```

```python
from sklearn.model_selection import train_test_split
def load_data(data, sentiments):
    global tokenizer
    indices = []
    for text in data:
      ids, segments = tokenizer.encode(text, max_len=SEQ_LEN)
      indices.append(ids)

    return [indices, np.zeros_like(indices)], np.array(sentiments)
```

```python
model.summary()
```

```python
data_train,data_test,label_train,label_test = train_test_split(data,label,test_size=0.2,random_state=42)
data_train,data_valid,label_train,label_valid = train_test_split(data_train,label_train,test_size=0.25,random_state=42)
```

```python
X_train,Y_train = load_data(data_train,label_train)
X_valid,Y_valid = load_data(data_valid,label_valid)
```

```python
X_test,Y_test = load_data(data_test,label_test)
```

```python
from keras.callbacks import ModelCheckpoint
checkpoint = ModelCheckpoint('best_model_mtl_BERT_1.h5', monitor='val_loss', #verbose=1,
                             save_best_only=True, mode='min')
callbacks_list = [checkpoint]
```

```python
model.fit(X_train,Y_train, epochs=EPOCHS, batch_size=BATCH_SIZE, verbose = 1, validation_data = (X_valid,Y_valid), callbacks=callbacks_list) 
```

```python
model.load_weights('best_model_mtl_BERT_1.h5')
```

```python
y_pred = np.round(model.predict(X_test))
from sklearn.metrics import classification_report
print(classification_report(Y_test, y_pred,digits=4))
```
